```python
import collections
import xlsxwriter
import datetime
import codecs
```

```python
def add_months(s, p):
    ''' 在日期s上加上p个月，如果日期不存在，顺延到下个月1号 '''
    p_years = s.year + p / 12
    p_months = s.month + p % 12 
    if p_months > 12:
        p_years += 1
        p_months = p_months % 12
    
    try:
        rs = datetime.datetime(p_years, p_months, s.day)
    except ValueError, e:
        if p_months == 12:
            rs = datetime.datetime(p_years + 1, 1, 1)
        else:
            rs = datetime.datetime(p_years, p_months+1, 1)
    return rs

def next_quater(cur_quater):
    return add_months(cur_quater, 3)

# 季度的分割日期，[datetime.datetime(2004, 1, 1, 0, 0), datetime.datetime(2004, 4, 1, 0, 0), datetime.datetime(2004, 7, 1, 0, 0) 。。。
quater_date = [datetime.datetime(2003,7,1),datetime.datetime(2003,10,1)] \
        + [datetime.datetime(v1,v2,1) for v1 in xrange(2004, 2018) for v2 in xrange(1,12,3)]
    
def rankname_2_score(rankname):
    if rankname == u'卖出':
        return 1
    elif rankname == u'减持':
        return 2
    elif rankname == u'中性':
        return 3
    elif rankname == u'增持':
        return 4
    elif rankname == u'买入':
        return 5
    else:
        raise ValueError
        
def mean(rank_list):
    return sum(rank_list) / float(len(rank_list))
```

```python
g_rank_data = collections.defaultdict(lambda :collections.defaultdict(list))
def load_data(file_name="AF_Bench_2002_2010.txt"):
    with codecs.open(file_name, "r", encoding="utf_16") as ifid:
        line_num = 0
        for line in ifid:
            if line_num > 0 and len(line) > 3:
                try:
                    Stkcd, Rptdt, Stdrank = line.strip().split('\t')
                    stock = Stkcd
                    Rptdt = datetime.datetime.strptime(Rptdt, "%Y-%m-%d")
                    quarter = datetime.datetime(Rptdt.year, (Rptdt.month-1)/3*3+1, 1)
                    score = rankname_2_score(Stdrank)
                    #print stock, quarter, score
                    if quarter in quater_date:
                        g_rank_data[stock][quarter].append(score)
                except ValueError, e:
                    print u"error at %s linenum %d: %s"  %(file_name, line_num+1, line.strip())
                
            line_num += 1
                
load_data("AF_Bench_2002_2010.txt")
load_data("AF_Bench_2010_2018.txt")
```

```python
with xlsxwriter.Workbook("rank_change_rate.xlsx") as wb:
    ws = wb.add_worksheet()
    ws.write(0,0,"stock")
    ws.write(0,1,"quarter")
    ws.write(0,2,"change rate")
    wt_idx = 1
    for fund_ in g_rank_data:
        start_date_ = sorted(g_rank_data[fund_].keys())[0]
        end_date_ = sorted(g_rank_data[fund_].keys())[-1]
        for i in xrange(quater_date.index(start_date_), quater_date.index(end_date_)+1):
            previous_rank_mean = mean(g_rank_data[fund_][quater_date[i-1]]) \
                                    if quater_date[i-1] in g_rank_data[fund_] else 0
            current_rank_meam = mean(g_rank_data[fund_][quater_date[i]]) \
                                    if quater_date[i] in g_rank_data[fund_] else 0
            
            if previous_rank_mean != 0 and current_rank_meam != 0:
                change_rate = (current_rank_meam - previous_rank_mean) / previous_rank_mean
            else:
                change_rate = 0
            ws.write(wt_idx,0,fund_)
            ws.write(wt_idx,1,u"%dQ%d"%(quater_date[i].year, (quater_date[i].month+2)/3))
            ws.write(wt_idx,2,change_rate)
            
            wt_idx += 1
```
